## **Handling Missing Value**
We will replace the missing values of country with mode

```python
df['country']=df['country'].fillna(df['country'].mode()[0])
df['country'].isnull().sum()
```

# **Data Cleaning**

```python
df.dropna(inplace=True)

# Drop Duplicates
df.drop_duplicates(inplace= True)
```

```python
df = df.drop(columns=['cast', 'show_id', 'title', 'description'],axis=1)
```

```python
df.head()
```

```python
df.rename(columns={'listed_in' :'genre'}, inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', infer_datetime_format=True)
```

# **WordCloud for every column**
WordCloud is the visual representation of the text data.Words are usually single and the fontsize or colour shows the importance of each word.

```python
from wordcloud import WordCloud
from PIL import Image
```

## **Directors**

```python
plt.subplots(figsize=(15,10))
text=" ".join(df.director.astype(str))
wordcloud = WordCloud(
                          margin=0,
                          background_color='white',
                          colormap='afmhot',
                          width=1800,
                          height=1000
                         ).generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.savefig('director.png')
plt.margins(x=0,y=0)
plt.show()
```

## **Country**
